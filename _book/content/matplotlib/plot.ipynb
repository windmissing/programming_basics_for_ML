# 绘制折线图

```python
import matplotlib.pyplot as plt
import numpy as np
```

# 绘制折线图

```python
dataSet = np.array([[1,2],   # 必须是np.array类型，普通的数组类型不可以
          [2,1],
          [3,4],
          [4,3]])
```

# plot的参数是两个一维数组

```python
plt.plot(dataSet[:,0], dataSet[:,1])   # 参数为坐标
plt.show()
```

# plot的参数是一个二维数组

```python
print(dataSet.shape)
plt.plot(dataSet)   # 参数为坐标
plt.show()
```

x轴的取值为[0, m)，绘制n根折线,第j根折线的第i个点为dataSet[i, j]
