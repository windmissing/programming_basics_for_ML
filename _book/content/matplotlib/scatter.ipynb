# 绘制散点图

```python
import matplotlib.pyplot as plt
import numpy as np
```

# 绘制散点图

```python
dataSet = np.array([[1,1],   # 必须是np.array类型，普通的数组类型不可以
          [2,2],
          [3,3],
          [4,4]])
```

```python
# 方法一
plt.scatter(dataSet[:,0], dataSet[:,1])   # 参数为坐标
plt.show()
```

```python
# 方法二
fig = plt.figure()
ax = fig.add_subplot(111)   # If no positional arguments are passed, defaults to (1, 1, 1)或(111)
ax.scatter(dataSet[:,0], dataSet[:,1])
plt.show()
```

# 绘制不同大小的散点图，以方法一为例，方法二类似

```python
# 参数1，2是坐标，参数3是大小。大小乘以15是为了让效果更明显
plt.scatter(dataSet[:,0], dataSet[:,1], s=15*np.array(dataSet[:,0]))
plt.show()
```

# 绘制不同颜色的散点图，以方法一为例，方法二类似

```python
# 参数1，2是坐标，参数3是颜色，此时必须标明"c="
# plt.scatter(x,y,c)
# 参数1，2是坐标，参数3是大小，参数4是颜色，此时不需要标明参数名
# plt.scatter(x,y,s,c)

# 颜色*15是为了让效果更明显
plt.scatter(dataSet[:,0], dataSet[:,1], c=15*np.array(dataSet[:,0]))
plt.show()
```

# 绘制圆环

```python
import matplotlib.pyplot as plt
# 0,0代表坐标
# s=150是大小
# c代表填充颜色，如果c='none'表示无填充
# edgecolor代表边框颜色，默认没有颜色。
# 无填充颜色+有边框颜色就是圆环的效果
# linewidth代表边框的粗细程度
plt.scatter(0,0, s=150, c='none', alpha=0.7, linewidth=1.5, edgecolor='#AB3319')
plt.show()
```

# 不同的形状

```python
import matplotlib.pyplot as plt
# 0,0代表坐标
# s=150是大小
# c代表填充颜色，如果c='none'表示无填充
# edgecolor代表边框颜色，默认没有颜色。
# 无填充颜色+有边框颜色就是圆环的效果
# linewidth代表边框的粗细程度
plt.scatter(0,0, s=150, marker='+')
plt.show()
```
