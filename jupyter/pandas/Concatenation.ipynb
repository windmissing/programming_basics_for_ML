```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame(np.ones((3,4)), columns=['a', 'b', 'c', 'd'])
df1
```

```python
df2 = pd.DataFrame(np.ones((3,4))*2, columns=['a', 'b', 'c', 'd'])
df2
```

```python
df3 = pd.DataFrame(np.ones((3,4))*3, columns=['a', 'b', 'c', 'd'])
df3
```

# 直接合并

```python
pd.concat([df1, df2, df3], axis=0)  # 索引号也直接合并
```

```python
pd.concat([df1, df2, df3], axis=0, ignore_index=True)  # 索引号重新生成
```

# data联结

```python
df1 = pd.DataFrame(np.ones((3,4)),
                   columns=['a', 'b', 'c', 'd'],
                   index=[1,2,3])
df1
```

```python
df2 = pd.DataFrame(np.ones((3,4))*2,
                   columns=['b', 'c', 'd', 'e'],
                   index=[2,3,4])
df2
```

```python
# 默认：join = 'outer'
# 上下合并时，列按列索引对齐

pd.concat([df1, df2])
```

```python
# join = 'inner': 只保留df1和df2共有的列

pd.concat([df1, df2], join='inner')
```

```python
pd.concat([df1, df2],axis=1, join_axes=[df1.index])
```

# append

```python
# append另一个data frame
df1.append(df2, ignore_index=True)
```

```python
# append一条数据
s = pd.Series([1,2,3,4], index=['a', 'b', 'c', 'd'])
df1.append(s, ignore_index=True)
```

```python
# append多条数据
df1.append([df2, df3], ignore_index=True)
```

# Merge

```python
left = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                     'A':['A0', 'A1', 'A2', 'A3'],
                     'B':['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key':['K0', 'K1', 'K2', 'K3'],
                      'C':['C0', 'C1', 'C2', 'C3'],
                      'D':['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on='key')
```

```python
# 两组Key
left = pd.DataFrame({'key1':['K0', 'K0', 'K1', 'K2'],
                     'key2':['K0', 'K1', 'K0', 'K1'],
                     'A':['A0', 'A1', 'A2', 'A3'],
                     'B':['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1':['K0', 'K1', 'K1', 'K2'],
                      'key2':['K0', 'K0', 'K0', 'K0'],
                      'C':['C0', 'C1', 'C2', 'C3'],
                      'D':['D0', 'D1', 'D2', 'D3']})
```

```python
# 默认的合并为inner，只有具有相同key的项才会保留
pd.merge(left, right, on=['key1','key2'])
```

```python
pd.merge(left, right, on=['key1','key2'], how='outer')
```

```python
# 基于第二组数据的key来合并
pd.merge(left, right, on=['key1','key2'], how='right')
```

```python
# indicator：显示merge数据的来源
pd.merge(left, right, on=['key1','key2'], how='outer', indicator=True)
```

```python
# indicator：显示merge数据的来源，并给这一列命名
pd.merge(left, right, on=['key1','key2'], how='outer', indicator='indicator_column')
```

# merge by index

```python
left = pd.DataFrame({'A':['A0', 'A1', 'A2'],
                     'B':['B0', 'B1', 'B2']},
                    index = ['K0', 'K1', 'K2'])
right = pd.DataFrame({'C':['C0', 'C1', 'C2'],
                      'D':['D0', 'D1', 'D2']},
                     index = ['K0', 'K2', 'K3'])
pd.merge(left, right, left_index=True, right_index=True, how='outer')
```

```python
pd.merge(left, right, left_index=True, right_index=True, how='inner')
```

# handle Overlapping

```python
# 两组数据有相同的列index，要把它们区分开
boys = pd.DataFrame({'k':['K0', 'K1', 'K2'],
                     'age':[1,2,3]})
girls = pd.DataFrame({'k':['K0', 'K0', 'K3'],
                      'age':[4,5,6]})
pd.merge(boys, girls, on='k', suffixes=['_boys', '_girls'], how='inner')
```
