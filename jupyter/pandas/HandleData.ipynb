```python
import pandas as pd
```

```python
df = pd.DataFrame()
df['t'] = [x for x in range(10)]
df
```

```python
# 默认生成浮点型数据
# 默认向下移动，上面补Nan
df['t-1'] = df['t'].shift(1)
df
```

```python
# 向上移动，则后面补Nan
df['t+1'] = df['t'].shift(-1)
df
```
