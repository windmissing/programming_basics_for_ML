# 随机剪裁

```python
import cv2
from matplotlib import pyplot as plt
image = cv2.imread("file/0a1ae83cd2072ca9168b7a17ea138b00.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
from torchvision.transforms import Compose, RandomCrop
from PIL import Image
compose = Compose([RandomCrop(100, 100)])  # 剪裁大小为100*100
image2 = compose(Image.fromarray(image))
plt.imshow(image2)
```

# 随机左右翻转

```python
from torchvision.transforms import Compose, RandomHorizontalFlip
from PIL import Image
compose = Compose([RandomHorizontalFlip()])  # 剪裁大小为100*100
image2 = compose(Image.fromarray(image))
plt.imshow(image2)
```
